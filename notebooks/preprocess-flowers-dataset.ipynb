# Preprocess Oxford Flowers Dataset

```python
import cv2
import os
import PIL

from typing import Dict, List

import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
```

```python
base_path = '../../datasets/flowers/102flowers/images'
os.listdir(base_path)
```

## Sample images

```python
sample_folder = f'{base_path}/train/1'
images_paths = [os.path.join(sample_folder, filename) for filename in os.listdir(sample_folder)]
images = [cv2.imread(image_path) for image_path in images_paths]

samples_overview = pd.DataFrame({
    'image_name': [image_path.split('/')[-1] for image_path in images_paths],
    'height': [image.shape[0] for image in images],
    'width': [image.shape[1] for image in images],
    'depth': [image.shape[2] for image in images]
})

samples_overview
```

```python
for image in images:
    plt.imshow(image)
    plt.show()
```

## Image preprocess

```python
size = 800 # Picked from EDA notebook

image_generator = ImageDataGenerator(rescale=1.0 / 255, fill_mode='nearest')
for _ in image_generator.flow_from_directory(directory=f'{base_path}/train', 
                                             target_size=(size,size), 
                                             save_to_dir='../../datasets/flowers/102flowers/preprocessed2/train'):
    pass
```
